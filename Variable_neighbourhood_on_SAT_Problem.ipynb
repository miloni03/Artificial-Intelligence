# **IMPLEMENTING VARIABLE NEIGHBOURHOOD ON SAT PROBLEM**

```python
from itertools import combinations
```

### **F1 = (A V ~ B) ^ (B V ~ C) ^ (~ B) ^ (~ C V E) ^ (A V C) ^ (~ C V ~ D)**

```python
def F1(values):
    A, B, C, D, E = values
    result = (A or not B) + (B or not C) + (not B) + (not C or E) + (A or C) + (not C or not D)
    return result
```

### **F2 =  (A V B) ^ (A ^ ~ C) ^ (B ^ D) ^ (A V ~ E)**

```python
def F2(values):
    A, B, C, D, E = values
    result = (A or B) + (A or not C) + (B or D) + (A or not E)
    return result
```

### **MOVE GEN FUNCTION**

```python
def swap(bit):
    if bit == 0:
        return 1
    return 0
```

```python
def moveGen(node, n):
    
    stateCopy = node.copy()
    pA, pB, pC, pD, pE = stateCopy
    parentF1 = F1(stateCopy)


    nextMoves = []
    index = [0,1,2,3,4]
    nextFlips = combinations(index, n)
    nextFlips = [list(i) for i in nextFlips]
    # print(len(nextFlips))
    for ls in nextFlips:

        temp = stateCopy.copy()
        
        for inx in ls:
            toFlip = temp[inx]
            temp[inx] = swap(toFlip)
            
        if F1(temp) > parentF1:
            # print(ls, temp, F1(temp))
            nextMoves.append([temp, F1(temp)])

    for i in nextMoves:
        print(i)

    return nextMoves
```

### **VND**

```python
def HC(START, n):
    OPEN = [[START, F1(START)]]
    CLOSED = []
    GOAL = 6
    depth = 0
    print('\n\nINITIAL OPEN:', OPEN)
    while OPEN!=[]:
        newParent = OPEN.pop(0)
        # print(newParent)
        print('\n\nCURRENT DEPTH:', depth)
        print('\nSELECTED:',newParent[0], '->', F1(newParent[0]))

        if F1(newParent[0]) == GOAL:
            print(f'SATISFIED AT DEPTH {depth} FOR:', newParent[0], '\n\n')
            print('-x-x-x-x-x-x-x-x-x-x-')
            return [True, START, depth]
        
        CLOSED.append(newParent)
        childNodes = moveGen(newParent[0], n)
        if(childNodes == []):
            print('NOT SATISFIED FOR:', START, 'XXXXX')
            return [False, START]

        print('\nCHILDREN OF',newParent[0])
        for child in childNodes:
            print(child[0],'->', child[1])
            OPEN.append(child)
        
        print('\n')
        depth += 1

        # OPEN.extend(childNodes)
        # print(OPEN)
        print('\nNEW OPEN:', OPEN,'\n')
        OPEN = sorted(OPEN, reverse = True, key = lambda x:x[1])
        OPEN = [OPEN[0]]
        print('\nNEW OPEN:', OPEN,'\n')
        # print('\n')

        print('----------')
        # break

    return [False, START]
```

```python
def VND(START):
    bestStart = None
    bestDepth = None
    bestFlips = None
    index = [0,1,2,3,4]
    for i in range(1,6):
        start = START.copy()
        print('\n\n=========================\n\n')
        print(f'FOR FLIPING {i} BITS')
        STARTS = []
        nextFlips = combinations(index, i)
        nextFlips = [list(i) for i in nextFlips]
        for ls in nextFlips:
            temp = start.copy()
            for inx in ls:
                toFlip = temp[inx]
                temp[inx] = swap(toFlip)
            STARTS.append(temp)

        for start in STARTS:
            print('\nFOR START =', start)
            newStart = HC(start, i)
            if(newStart[0]):
                if(bestStart == None) or (newStart[2] < bestDepth):
                    bestStart = newStart[1]
                    bestDepth = newStart[2]
                    bestFlips = 1

            print(f'\n\nTILL BEST START: {bestStart} FOR {bestFlips} NUMBER OF FLIPS, WITH MAX DEPTH = {bestDepth}')
            print('\n\n=========================\n\n')
    
    return [bestStart, bestDepth]
```
